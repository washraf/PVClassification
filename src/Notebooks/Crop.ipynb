```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
```

```python
def SaveImage(imgf,name):
    fig = plt.figure(figsize = [25,25])
    plt.imshow(cv2.cvtColor(imgf, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    #plt.show()
    fig.savefig(name, bbox_inches='tight', pad_inches=-0.5)
    plt.close()
```

```python
import os
import fnmatch
def files_within(directory_path, pattern="*"):
    for dirpath, dirnames, filenames in os.walk(directory_path):
        for file_name in fnmatch.filter(filenames, pattern):
            yield os.path.join(dirpath, file_name)
```

```python
def cropeImage(img):
    grayed = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    (_,thresh) = cv2.threshold(grayed,150,255,cv2.THRESH_BINARY)
    result, contours, _= cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    x, y = [], []
    for contour_line in contours:
        for contour in contour_line:
            x.append(contour[0][0])
            y.append(contour[0][1])
    x1, x2, y1, y2 = min(x), max(x), min(y), max(y)
    cropped = img[y1:y2, x1:x2]
    return cropped,thresh
```

```python
def computeandSave(file_path):
    img = cv2.imread(file_path)
    #img = RemoveBG(img)
    img,_ = cropeImage(img)
    name = file_path.split('\\')[-1]
    #print(name)
    rest = file_path.split('D:\\Solar Panels\\NoBG\\')[-1]
    #print(rest)
    directory = "D:\\Solar Panels\\Cropped\\"+rest.split(name)[0]
    if not os.path.exists(directory):
        os.makedirs(directory)
    SaveImage(img,directory+name)
```

```python
images = list(files_within("D:\\Solar Panels\\NoBG\\","*.JPG"))
print(len(images))
for file_path in images:
    try:
        print(file_path + " Start !")
        computeandSave(file_path)
    except:
        print(file_path + " is error !")
```
