```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
```

```python
#plt.imshow(cv2.cvtColor(imgo, cv2.COLOR_BGR2RGB))
#plt.show()
```

```python
def RemoveBG(imgo):
    #print("Remove BG")
    mask = np.zeros(imgo.shape[:2],np.uint8)
    bgdModel = np.zeros((1,65),np.float64)
    fgdModel = np.zeros((1,65),np.float64)
    rect = (50,50,2300,2000)
    cv2.grabCut(imgo,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
    mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
    imgf = imgo*mask2[:,:,np.newaxis]
    return imgf
```

```python
def SaveImage(imgf,name):
    fig = plt.figure(figsize = [25,25])
    plt.imshow(cv2.cvtColor(imgf, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    #plt.show()
    fig.savefig(name, bbox_inches='tight', pad_inches=-0.5)
    plt.close()
```

```python
import os
import fnmatch
def files_within(directory_path, pattern="*"):
    for dirpath, dirnames, filenames in os.walk(directory_path):
        for file_name in fnmatch.filter(filenames, pattern):
            yield os.path.join(dirpath, file_name)
```

```python
def computeandSave(file_path):
    img = cv2.imread(file_path)
    img = RemoveBG(img)
    name = file_path.split('\\')[-1]
    #print(name)
    rest = file_path.split('F:\\Solar Data\\')[-1]
    #print(rest)
    directory = "F:\\NoBG\\"+rest.split(name)[0]
    if not os.path.exists(directory):
        os.makedirs(directory)
    SaveImage(img,directory+name)
```

```python
images = list(files_within("F:\\Solar Data\\27-8-18","*.JPG"))
print(len(images))
for file_path in images:
    print(file_path + " Start !")
    computeandSave(file_path)
    #print(file_path + " is done !")
```
