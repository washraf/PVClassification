# Imports

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import os, os.path
import numpy as np
from scipy import misc
import pandas as pd
from skimage import color
#%pylab inline
```

```python
ImgsType = "Clean"
imgs = []
path = "../Images/"+ImgsType
valid_images = [".jpg",".png"]
for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_images:
        continue
    imgs.append(misc.imread(os.path.join(path,f)))
```

# Create Data Set

```python
data = []
for i in range(0,len(imgs)):
    A = imgs[i]
    R = imgs[i][0]
    G = imgs[i][1]
    B = imgs[i][2]
    Gray = color.rgb2gray(imgs[i])
    data.append([ A.mean(),A.std(),  R.mean(),R.std(),  G.mean(),G.std(),  B.mean(),B.std(), G.mean(),G.std() ])
numpyarray = np.array(data)
```

```python
numpyarray.shape
```

```python
df = pd.DataFrame(numpyarray, columns = ["Amean","Avar","Rmean","Rvar","Gmean","Gvar","Bmean","Bvar","Kmean","Kvar",])
```
